# ***Multiple Filter Criteria to a pandas DataFrame***

```python
import pandas as pd #importing module
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies[movies.duration>=200]
```

```python
# movies that are 200 minutes long but also of Drama genre & for and and | for or
movies[(movies.duration>=200) & (movies.genre=='Drama')]
```

```python
movies.duration >= 200
```

```python
movies.genre == 'Drama'
```

```python
movies[(movies.duration>=200)&(movies.genre == 'Drama')]
```

```python
#  how to get movies with adventure and drama genere and Crime

movies[movies.genre.isin(['Drama', 'Crime', 'Adventure'])]
```
